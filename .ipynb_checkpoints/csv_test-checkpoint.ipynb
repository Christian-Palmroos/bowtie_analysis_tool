```python
import Bowtie

import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
```

```python
data_file_name = "array_vault_e_256"

# The number of particles shot in a simulation of all energy bins
particles_shot = np.load(f'{data_file_name}/particles_Shot.npy')

# The number of particles detected per particle channel in all energy bins
particles_response = np.load(f'{data_file_name}/particles_Respo.npy')

other_params = np.load(f'{data_file_name}/other_params.npy')

# The total number of energy bins
nstep = int(other_params[0])

# The radiation area (isotropically radiating sphere) around the Geant4 instrument model in cm2
radiation_area = other_params[2]

# Midpoints of the energy bins in MeV
energy_midpoint = np.load(f'{data_file_name}/energy_Mid.npy')

# High cuts of the energy bins in MeV
energy_toppoint = np.load(f'{data_file_name}/energy_Cut.npy')

# The energy bin widths in MeV
energy_channel_width = np.load(f'{data_file_name}/energy_Width.npy')
```

```python
# energy grid in the format compatible with the output of a function in the bowtie package
energy_grid = { 'nstep': nstep,
                'midpt': energy_midpoint,
                'ehigh': energy_toppoint, 
                'enlow': energy_toppoint - energy_channel_width,
                'binwd': energy_channel_width }
```

```python
import pandas as pd

#from Bowtie import Bowtie, Spectra
import Bowtie as bow
```

```python
response_df = pd.read_csv("sixs_side0_electron_responses.csv", index_col="incident_energy")
display(response_df)
```

```python
# Initialize a range of spectra, starting from minimum to maximum with an amount of steps
spectra = bow.Spectra(gamma_min=-3.5, gamma_max=-1.5, gamma_steps=100)
```

```python
# Produces the power law spectra with given spectral indices for the energies 
# in the response dataframe indices
spectra.produce_power_law_spectra(response_df)
```

```python
# Initializes the bowtie object with an energy range (in MeV) and the response data
bowtie = bow.Bowtie(energy_min=0.01, energy_max=50., data=response_df)
```

```python
# bowtie_analysis is a class method for a single channel bowtie analysis with the given spectra
channel_results = bowtie.bowtie_analysis(channel="E1", spectra=spectra, plot=True)
```

```python
# The returned channel_results is a dictionary that contains:
# geometric_factor : {float} The geometric factor G\DeltaE in cm^2 sr MeV
# geometric_factor_errors (absolute) : {dict} with 'gfup' : {float}, and 'gflo' : {float}
# effective_energy : {float} The effective energy in MeV
# effective_lower_boundary : {float} 
# effective_upper_boundary : {float}
# fig : {plt.Fig} if plot==True. The figure that visualizes the spectra and the response function
# ax : {tuple(plt.Axes)} if plot==True. The two pairs of axes of the figure.
channel_results
```

```python
all_channel_results = bowtie.bowtie_analysis_full_stack(spectra=spectra, plot=True)
```

```python
for res in all_channel_results:
    print(f"G: {res['geometric_factor']:.3g} cm2srMeV")
```

```python
import numpy as np
import matplotlib.pyplot as plt

box_response = np.ones(len(response_df.index))

low_bound = 0.57
high_bound = 2.57
gdeltae = 0.017

for i, resp in enumerate(box_response):
    box_response[i] = resp * gdeltae if response_df.index.values[i] > low_bound and response_df.index.values[i] < high_bound else 0

fig, ax = plt.subplots(figsize=(13,7))

ax.set_yscale("log")
ax.set_xscale("log")
ax.step(response_df.index, box_response)

plt.show()
```

```python
def save_results(results, column_names=None, savefigs=False):
    """
    Saves the results Bowtie.bowtie_analysis or Bowtie.bowtie_analysis_full_stack
    
    Parameters:
    -----------
    results : {list[dict]}
    
    column_names : {list[str]}, optional
    
    savefigs : {bool} Saves the figures in .png format.
    """

    # The indices (rows) of the output file
    INDICES = ["geometric_factor", "gf_error_up", "gf_error_low", "effective_energy"]

    # Make sure that results are a list of dictionaries to iterate over
    if isinstance(results,dict):
        results = [results]

    # Handle column names. If single string, make sure it's contained in a list.
    # If not given, procude a range of integers for placeholders
    if isinstance(column_names,str):
        column_names = [column_names]

    if column_names is None:
        columns_names = range(len(results))

    data = np.empty((len(INDICES),len(results)))
    # Loop through the list of results:
    for i, res in enumerate(results):
        
        data[0,i] = res["geometric_factor"]
        data[1,i] = res["geometric_factor_errors"]["gfup"]
        data[2,i] = res["geometric_factor_errors"]["gflo"]
        data[3,i] = res["effective_energy"]
    
    result_df = pd.DataFrame(data=data, index=INDICES, columns=column_names)
    return result_df

d = save_results(results=all_channel_results, )
d
```
